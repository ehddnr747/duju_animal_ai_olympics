```python
def moving_average(l, size):
    r = []
    
    for i in range(1, len(l)+1):
        if i < size:
            r.append(np.sum(l[0:i]) / i)
        else :
            r.append(np.sum(l[i-size:i])/size)
    
    return r
        
```

```python
prop_exp_dirs = sorted(glob.glob("/home/duju/animal_ai_olympics/duju_animal_ai_olympics/proposed_results/*"))
exp_dirs = [
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline2_1-Food",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline4_2-Preferences",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline3_3-Obstacles",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline4_4-Avoidance",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline2_5-SpatialReasoning",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline4_6-Generalization",
    "/home/duju/animal_ai_olympics/duju_animal_ai_olympics/baseline_results/Animal_AI_Baseline3_7-InternalMemory"
]

base_plot_dict = {}
prop_plot_dict = {}

for exp_num in range(1, 8):
    for exp_dir in [i for i in exp_dirs if str(exp_num)+"-" in i]:
        with open(os.path.join(exp_dir,"results.txt"), "r") as f:
            txts = f.readlines()
        txts = [i for i in txts if "***" in i]
        
        if exp_num == 2 :
            txts = [float(i.split("***")[2]) > 0 and int(i.split("***")[1]) < 251 for i in txts]
        else :
            txts = [float(i.split("***")[2]) > 0 for i in txts]
        txts = moving_average(txts, 100)
        base_plot_dict[exp_num] = np.array(txts)
    
    
    for exp_dir in [i for i in prop_exp_dirs if str(exp_num)+"-" in i]:
        with open(os.path.join(exp_dir,"results.txt"), "r") as f:
            txts = f.readlines()
        txts = [i for i in txts if len(i.split("***"))==4]
        #assert len(txts) == 1000
        #txts = [float(i.split("***")[2]) > 0.0 and int(i.split("***")[1]) < 251 for i in txts]
        if exp_num == 2 :
            txts = [float(i.split("***")[2]) > 0 and int(i.split("***")[1]) < 251 for i in txts]
        else :
            txts = [float(i.split("***")[2]) > 0 for i in txts]
        txts = moving_average(txts, 100)
        prop_plot_dict[exp_num] = np.array(txts)
```

```python
fig = plt.figure(figsize=(32, 6))

fig.add_subplot(1, 3, 1).plot(prop_plot_dict[1], color = "steelblue", linewidth = 3)
fig.add_subplot(1, 3, 1).plot(base_plot_dict[1], color = "salmon", linewidth = 3)
fig.add_subplot(1, 3, 1).set_title("1-Food", fontsize = 20)
fig.add_subplot(1, 3, 1).set_ylabel("success rate for 100 episodes", labelpad=40, fontsize= 18)
fig.add_subplot(1, 3, 1).tick_params(axis='both', which='major', labelsize=15)

fig.add_subplot(1, 3, 2).plot(prop_plot_dict[2], color = "steelblue", linewidth = 3)
fig.add_subplot(1, 3, 2).plot(base_plot_dict[2], color = "salmon", linewidth = 3)
fig.add_subplot(1, 3, 2).set_title("2-Preferences", fontsize = 20)
fig.add_subplot(1, 3, 2).tick_params(axis='both', which='major', labelsize=15)

fig.add_subplot(1, 3, 3).plot(prop_plot_dict[3], color = "steelblue", linewidth = 3)
fig.add_subplot(1, 3, 3).plot(base_plot_dict[3], color = "salmon", linewidth = 3)
fig.add_subplot(1, 3, 3).set_title("3-Obstacles", fontsize = 20)
fig.add_subplot(1, 3, 3).tick_params(axis='both', which='major', labelsize=15)

plt.suptitle("success rate curve for 2.5e5 iterations", x=0.45, y=0.05, horizontalalignment='left', verticalalignment='top', fontsize = 18)

l1, l2 = fig.add_subplot(1, 3, 1).bar([0,0],[0,0],color=("steelblue","salmon"))
plt.figlegend((l1, l2),["Proposed","Baseline"], "upper center",ncol=2, fontsize=15, borderaxespad=0.0)

plt.savefig("success_curve_1.png")
```
