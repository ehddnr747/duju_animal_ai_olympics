```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("./log/Rainbow Double-PER-Duel DQN -Animal Obstacles","r") as f:
    txts = f.readlines()
```

```python
txts = [i for i in txts if "\t" in i]
```

```python
rewards = [float(i.split("*")[6].strip()) for i in txts]
```

```python
moving_rewards = []
for i in range(10, len(rewards)):
    moving_rewards.append( np.mean(rewards[i-10:i]))
```

```python
plt.plot(moving_rewards)
```

```python
success = [float(i.split("*")[6].strip()) >= 0.0 for i in txts]
```

```python
moving_success = []
for i in range(100, len(success)):
    moving_success.append( np.mean(success[i-100:i]))
```

```python
plt.plot(moving_success)
```
