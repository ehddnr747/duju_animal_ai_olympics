```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn
```

```python
with open("../log/ddqn frame 1 good but still diverge","r") as f:
    lines = f.readlines()
```

```python
tuples = []
for j in [i.split("*** ")[1].strip().split(",") for i in lines if "***" in i and "Eval" not in i]:
    tuples.append((float(j[0][1:].strip()), 
                   float(j[1].strip()), 
                   float(j[2].strip()), 
                   float(j[3].strip()), 
                   float(j[4][:-1].strip())))
```

```python
len(tuples)
```

```python
q1_mean = np.array([i[0] for i in tuples])
q1_min = np.array([i[1] for i in tuples])
q1_max = np.array([i[2] for i in tuples])
q2_mean = np.array([i[3] for i in tuples])
r_mean = np.array([i[4] for i in tuples])
```

```python
plt.bar(range(len(q1_mean)),q1_mean,width=1.0)
plt.xlabel("episodes")
plt.ylabel("mean q1")
plt.savefig("diverge_mean_q1.png")
plt.show()

plt.bar(range(len(q1_mean)),q1_min,width=1.0)
plt.xlabel("episodes")
plt.ylabel("min q1")
plt.savefig("diverge_min_q1.png")
plt.show()

plt.bar(range(len(q1_mean)),q1_max,width=1.0)
plt.xlabel("episodes")
plt.ylabel("max q1")
plt.savefig("diverge_max_q1.png")
plt.show()

plt.bar(range(len(q1_mean)),q2_mean,width=1.0)
plt.xlabel("episodes")
plt.ylabel("q2 reward")
plt.savefig("diverge_mean_q2.png")
plt.show()

plt.bar(range(len(q1_mean)),r_mean,width=1.0)
plt.xlabel("episodes")
plt.ylabel("mean reward")
plt.savefig("diverge_mean_reward.png")
plt.show()
```
