```python
def moving_average(l, size):
    r = []
    
    for i in range(1, len(l)+1):
        if i < size:
            r.append(np.sum(l[0:i]) / i)
        else :
            r.append(np.sum(l[i-size:i])/size)
    
    return r
        
```

```python
with open("./log/Rainbow Double-PER-Duel DQN -Animal Obstacles","r") as f:
    txts = f.readlines()
```

```python
txts = [i for i in txts if "\t" in i]
```

```python
rewards = [float(i.split("*")[6].strip()) for i in txts]
```

```python
len(rewards)
```

```python
rewards = rewards[:1000]
```

```python
plt.plot(rewards)
plt.xlabel("episode")
plt.ylabel("reward")
plt.savefig("Rainbow Double-PER-Duel DQN -Animal Obstacles_rewards.png")
plt.show()
```

```python
moving_rewards = []
for i in range(100, len(rewards)):
    moving_rewards.append( np.mean(rewards[i-100:i]))
```

```python
plt.plot(moving_rewards)
plt.xlabel("episode")
plt.ylabel("reward of moving 100 episodes")
plt.savefig("Rainbow Double-PER-Duel DQN -Animal Obstacles_rewards_100_moving.png")
plt.show()
```

```python
success = [float(i.split("*")[6].strip()) >= 0.0 for i in txts]
```

```python
moving_success = []
for i in range(100, len(success)):
    moving_success.append( np.mean(success[i-100:i]))
```

```python
moving_success = moving_success[:1000]
```

```python
plt.plot(moving_success)
plt.xlabel("episode")
plt.ylabel("success rate of 100 episodes")
plt.savefig("Rainbow Double-PER-Duel DQN -Animal Obstacles_success_rate_100_moving.png")
plt.show()
```
